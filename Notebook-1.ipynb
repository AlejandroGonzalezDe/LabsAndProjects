```sql
select People.EmailAddress, People.FullName, People.LogonName, 
SUBSTRING(people.EmailAddress, 0, (CHARINDEX('@', People.EmailAddress))) as MailBox,
SUBSTRING(people.EmailAddress, (CHARINDEX('@', People.EmailAddress))+1,Len(People.EmailAddress)) as Domain,
People.IsEmployee,
People.IsSalesperson

from APplication.People People
where people.EmailAddress is not null;
```

```sql
select People.EmailAddress, 
SUBSTRING(people.FullName, 1, 1) + '. ' + SUBSTRING(people.FullName, (CHARINDEX(' ', People.FullName))+1,Len(People.FullName)) as Name, 
IIF(IsEmployee=1, 'Employee', 'SalesPerson') as Person_Type, 
Case when (IsEmployee=1) then 'Employee'
    when (IsSalesperson=1) then 'SalesPerson' 
    end ContactType
from APplication.People People 
where people.EmailAddress is not null;
```

```sql
select count(People.FullName) as total,
case 
when (IsSalesPerson> 0) then Count(IsSalesPerson) end as Sales_Person,
case 
when (IsEmployee> 0) then Count(IsSalesPerson) end as Employee,
case 
when (IsSystemUser> 0) then Count(IsSalesPerson) end as SystemUser,
case 
when (IsSystemUser is null and  IsEmployee is null and IsSalesPerson is null) then Count(People.FullName) end as OtherPerson
from Application.People


Group by IsSalesperson, IsEmployee, IsSystemUser;
```
